```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
x,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
x
```

```python
y
```

```python
pd.DataFrame(x)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel = 'linear')
svc.fit(x_train,y_train)
svc.coef_
```

```python
y_pred=svc.predict(x_test)
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

# RBF

```python
rbf = SVC(kernel = 'rbf')
rbf.fit(x_train,y_train)
y_pred_rbf=rbf.predict(x_test)
print(classification_report(y_test,y_pred_rbf))
print(confusion_matrix(y_test,y_pred_rbf))
```

# Polynomial

```python
polynomial = SVC(kernel = 'poly')
polynomial.fit(x_train,y_train)
y_pred_poly=polynomial.predict(x_test)
print(classification_report(y_test,y_pred_poly))
print(confusion_matrix(y_test,y_pred_poly))
```

# Sigmoid

```python
sigmoid = SVC(kernel = 'sigmoid')
sigmoid.fit(x_train,y_train)
y_pred_sigmoid=sigmoid.predict(x_test)
print(classification_report(y_test,y_pred_sigmoid))
print(confusion_matrix(y_test,y_pred_sigmoid))
```

# Hyper Parameter Tunning with SVC

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter range
 
param_grid={'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001],'kernel':['rbf']}
grid = GridSearchCV(SVC(),param_grid,refit=True,cv=5,verbose=3)
grid.fit(x_train,y_train)
grid.best_params_
```

```python

grid_predictions = grid.predict(x_test)
print(classification_report(y_test,grid_predictions))
print(confusion_matrix(y_test,grid_predictions))
```

# C = Regularization Parameter
# gama = Influence of single data point
Total combinations
5(c)*5(gama)*1(kernel) = 25 models

here we used GridSearchCV - used to perform hyperparameter tunning.

It checks all combinations of given parameters using cv and selects the model with best performance...
